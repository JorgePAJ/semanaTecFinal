```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#Cargar el archivo usando pandas
data = pd.read_csv("starbucks_drinkMenu_expanded.csv")
#Visualizar los datos cargados
data.head(20)
```

# Aqui comienza la realización de la actividad 2 - Obtención de metricas descriptivas

```python
#Definir la cantidad de datos en el archivo usado, asi como los tipos de variables con las que se esta lidiando
num_rows, num_cols = data.shape
data_types = data.dtypes

print(f"Row count: {num_rows}")
print(f"Column count: {num_cols}")
print("Variable and type:")
print(data_types)
```

```python
#Carga de los datos de cada columna numerica
category = data['Beverage_category']
calories = data['Calories']
trans_fat = data['Trans Fat (g) ']
satu_fat = data['Saturated Fat (g)']
sodium = data[' Sodium (mg)']
carbos = data[' Total Carbohydrates (g) ']
chol = data['Cholesterol (mg)']
fibre = data[' Dietary Fibre (g)']
sugar = data[' Sugars (g)']
protein = data[' Protein (g) ']
```

```python
#Resumen de estadisticas
cal = calories.describe()
trans = trans_fat.describe()
satu = satu_fat.describe()
sod = sodium.describe()
carbs = carbos.describe()
chols = chol.describe()
fib = fibre.describe()
sug = sugar.describe()
prot = protein.describe()
```

```python
#Obtencion de metricas por variable
media_cal = cal['mean']
median_cal = cal['50%']
std_cal = cal['std']

print("--- Calorias ---")
print(f"Media: {media_cal}")
print(f"Mediana: {median_cal}")
print(f"Desviacion: {std_cal}")
```

```python
media_trans = trans['mean']
median_trans = trans['50%']
std_trans = trans['std']

print("--- Trans Fats ---")
print(f"Media: {media_trans}")
print(f"Mediana: {median_trans}")
print(f"Desviacion: {std_trans}")
```

```python
media_satu = satu['mean']
median_satu = satu['50%']
std_satu = satu['std']

print("--- Saturated Fats ---")
print(f"Media: {media_satu}")
print(f"Mediana: {median_satu}")
print(f"Desviacion: {std_satu}")
```

```python
media_sod = sod['mean']
median_sod = sod['50%']
std_sod = sod['std']

print("--- Sodium ---")
print(f"Media: {media_sod}")
print(f"Mediana: {median_sod}")
print(f"Desviacion: {std_sod}")
```

```python
media_chols = chols['mean']
median_chols = chols['50%']
std_chols = chols['std']

print("--- Cholesterol ---")
print(f"Media: {media_chols}")
print(f"Mediana: {median_chols}")
print(f"Desviacion: {std_chols}")
```

```python
media_fib = fib['mean']
median_fib = fib['50%']
std_fib = fib['std']

print("--- Fibre ---")
print(f"Media: {media_fib}")
print(f"Mediana: {median_fib}")
print(f"Desviacion: {std_fib}")
```

```python
media_sug = sug['mean']
median_sug = sug['50%']
std_sug = sug['std']

print("--- Sugar ---")
print(f"Media: {media_sug}")
print(f"Mediana: {median_sug}")
print(f"Desviacion: {std_sug}")
```

```python
media_prot = prot['mean']
median_prot = prot['50%']
std_prot = prot['std']

print("--- Protein ---")
print(f"Media: {media_prot}")
print(f"Mediana: {median_prot}")
print(f"Desviacion: {std_prot}")
```
