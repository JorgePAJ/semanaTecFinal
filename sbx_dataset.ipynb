# Aqui comienza la realizacion de la actividad 3 - Mapas de color y boxplots


```python
fig = plt.figure(figsize=(25, 25))
gs = gridspec.GridSpec(3, 3)
fig.suptitle("Normal distribution for each numeric variable", fontsize=16)

ax1 = plt.subplot(gs[0, 0])  # First row, first column
ax2 = plt.subplot(gs[0, 1])  # First row, second column
ax3 = plt.subplot(gs[0, 2])  # First row, third column
ax4 = plt.subplot(gs[1, 0])  # Second row, first column
ax5 = plt.subplot(gs[1, 1])  # Second row, second column
ax6 = plt.subplot(gs[1, 2])  # Second row, third column
ax7 = plt.subplot(gs[2, 0])  # Third row, first column
ax8 = plt.subplot(gs[2, 1])  # Third row, second column
ax9 = plt.subplot(gs[2, 2])  # Third row, third column


sns.histplot(data=data, x='Calories', hue = 'Beverage_category', ax = ax1, bins=20, kde= True)
sns.histplot(data=data, x='Trans Fat (g) ', hue = 'Beverage_category', ax = ax2, bins=20, kde= True)
sns.histplot(data=data, x='Saturated Fat (g)', hue = 'Beverage_category', ax = ax3, bins=20, kde= True)
sns.histplot(data=data, x=' Sodium (mg)', hue = 'Beverage_category', ax = ax4, bins=20, kde= True)
sns.histplot(data=data, x=' Total Carbohydrates (g) ', hue = 'Beverage_category', ax = ax5, bins=20, kde= True)
sns.histplot(data=data, x='Cholesterol (mg)', hue = 'Beverage_category', ax = ax6, bins=20, kde= True)
sns.histplot(data=data, x=' Dietary Fibre (g)', hue = 'Beverage_category', ax = ax7, bins=20, kde= True)
sns.histplot(data=data, x=' Sugars (g)', hue = 'Beverage_category', ax = ax8, bins=20, kde= True)
sns.histplot(data=data, x=' Protein (g) ', hue = 'Beverage_category', ax = ax9, bins=20, kde= True)

ax9.set_title("Calories")
ax1.set_title("Trans Fats")
ax2.set_title("Saturated Fats")
ax3.set_title("Sodium")
ax4.set_title("Carbohydrates")
ax5.set_title("Cholesterol")
ax6.set_title("Fibre")
ax7.set_title("Sugars")
ax8.set_title("Protein")


plt.tight_layout()
plt.show()
```

```python
#Relacion numerica entre pares de variables numericas dividas en colores por categorias
sns.pairplot(data=data, hue='Beverage_category')
plt.suptitle('Relation between each pair of numeric variables', y=1.05)
```

```python
#Mapa de correlacion entre las variables numericas
sbx_corr = data.corr()
sns.heatmap(data=sbx_corr, vmin=-1, vmax=1, cmap = 'RdBu', annot=True, square = True)
plt.title('Heatmap of the correlation between variables')
```

```python
#Representacion grafica de los cuartiles de cada variable numerica divida en categorias
fig = plt.figure(figsize=(25, 10))
gs = gridspec.GridSpec(3, 3)
fig.suptitle("Quartile representations for each numeric variable", fontsize=16)

ax1 = plt.subplot(gs[0, 0])  # First row, first column
ax2 = plt.subplot(gs[0, 1])  # First row, second column
ax3 = plt.subplot(gs[0, 2])  # First row, third column
ax4 = plt.subplot(gs[1, 0])  # Second row, first column
ax5 = plt.subplot(gs[1, 1])  # Second row, second column
ax6 = plt.subplot(gs[1, 2])  # Second row, third column
ax7 = plt.subplot(gs[2, 0])  # Third row, first column
ax8 = plt.subplot(gs[2, 1])  # Third row, second column
ax9 = plt.subplot(gs[2, 2])  # Third row, third column


sns.boxplot(data=data, x='Calories', y = 'Beverage_category', ax = ax1)
sns.boxplot(data=data, x='Trans Fat (g) ', y = 'Beverage_category', ax = ax2)
sns.boxplot(data=data, x='Saturated Fat (g)', y = 'Beverage_category', ax = ax3)
sns.boxplot(data=data, x=' Sodium (mg)', y = 'Beverage_category', ax = ax4)
sns.boxplot(data=data, x=' Total Carbohydrates (g) ', y = 'Beverage_category', ax = ax5)
sns.boxplot(data=data, x='Cholesterol (mg)', y = 'Beverage_category', ax = ax6)
sns.boxplot(data=data, x=' Dietary Fibre (g)', y = 'Beverage_category', ax = ax7)
sns.boxplot(data=data, x=' Sugars (g)', y = 'Beverage_category', ax = ax8)
sns.boxplot(data=data, x=' Protein (g) ', y = 'Beverage_category', ax = ax9)


ax1.set_title("Calories")
ax2.set_title("Trans Fats")
ax3.set_title("Saturated Fats")
ax4.set_title("Sodium")
ax5.set_title("Carbohydrates")
ax8.set_title("Cholesterol")
ax8.set_title("Fibre")
ax8.set_title("Sugars")
ax9.set_title("Protein")


plt.tight_layout()
plt.show()
```
